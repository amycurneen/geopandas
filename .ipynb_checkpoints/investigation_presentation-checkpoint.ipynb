# The basics: importing & working with a world basemap

Geopandas comes with some default basemaps that we can use for analysis

```python
# this dataset contains a shape dataset for each country, as well as GDP and population of the country

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.head()
```

```python
ax = world.plot(figsize=(15,5))
```

```python
#remove antartica by name to center the map
world = world[(world.pop_est>0) & (world.name!="Antarctica") & (world.name!="Fr. S. Antarctic Lands")]
ax = world.plot(figsize=(15,5))
```

```python
# Let's create a new column named `gdp_per_cap` and calculate existing columns to calculate the GDP.

world['gdp_per_cap'] = world.gdp_md_est / world.pop_est
world.sort_values('gdp_per_cap', ascending = False).head()
world.head()
```

```python
#we can display this data on the map based on a color scale
base = world.plot(column='gdp_per_cap', cmap='OrRd', scheme='quantiles', figsize=(15,5))
plt.savefig("test.png")
```

```python
# remember- we're still in pandas, so we can do all the normal pandas things
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(world['gdp_per_cap'], bins=40)
ax.set_title("Histogram of GDP Per Capita")
```

## Dealing with projections

```python
## A note on projections
# projections are powerful- they can completely reshape your data, or completely mess up your workflow.

# When working with multiple datasets, always check that the projections match first

world.crs
```

```python
# It's WGS-84, a spherical coordinate system AKA 'epsg:4326'
# It can't be shown on a flat surface but when you try typically gets projected into a Platte Carree projection
# This the the CRS used by GPS/Satellites and GeoJSON
ax = world.plot()
ax.set_title("WGS84 (lat/lon)")
```

```python
# Set the projection to the Mercator projection (epsg=3395)
world = world.to_crs(epsg=3395)
ax = world.plot()
ax.set_title("Mercator")
```

```python
# Set the projection to Polar Stereographic (epsg=3995)
world = world.to_crs(epsg=3995)
ax = world.plot()
ax.set_title("Polar Stereographic")
```

# Mapping eclipses

**Data: **
Eclipse data downloaded by event in KMZ format here: http://xjubier.free.fr/en/site_pages/SolarEclipsesGoogleEarth.html

**Process: **
Used the Esri KML to Layer tool on each KMZ to export to a layer file.
Exported the polygon portion of said layer file to a shapefile
Used Esri's Merge(Data Management) tool to merge all the individual polygon shapefiles into one.


```python
#Read in Eclipse data
eclipses = gpd.read_file("Data/Eclipses.shp")
eclipses.head()
```

```python
# Check the coordinate reference system of our data - notice it oesn't match!
print(eclipses.crs,world.crs)
eclipses.crs == world.crs
```

```python
#fix the world basemap projection
world = world.to_crs(epsg=4326)
```

```python
ax = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,5))

# We added categorical, column, and legend properties
eclipses.plot(ax=ax, cmap='tab10', categorical=True, alpha=0.5, column = "Year", legend=True)
ax.axis('off');
```

## Adding in citites data

**Data: ** http://www.naturalearthdata.com/downloads/10m-cultural-vectors/10m-populated-places/

```python
#let's add in some citites data
cities = gpd.read_file("Data/ne_10m_populated_places.shp")
cities.head()
```

```python
#check the projections again
eclipses.crs == cities.crs
```

```python
#add the citites points to the map and view them
ax = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,5))
eclipses.plot(ax=ax, cmap='tab10', alpha=0.5)
cities.plot(marker='*', color='yellow', markersize=5, ax=ax)
ax.axis('off');
```

# Question: 
## Where should I travel to see an eclipse in the next 2 years?

```python
eclipses.Year.value_counts()
```

```python
#subset eclipses to get the next 2 years
myeclipse = eclipses[(eclipses['Year'] == (2019))| (eclipses['Year']== (2020))]
myeclipse
```

```python
#plot my eclipses and cities
ax = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,5))
myeclipse.plot(ax=ax)
cities.plot(marker='*', color='yellow', markersize=5, ax=ax)
ax.axis('off');
```

```python
# Style the plot so that the eclipses looks eclipse-ier
ax = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,10))
myeclipse.plot(ax=ax, color='yellow', edgecolor='black', alpha=0.5)
cities.plot(marker='*', color='blue', markersize=5, ax=ax)
bounds = myeclipse.geometry.bounds
plt.xlim([bounds.minx.min()-5, bounds.maxx.max()+5])
plt.ylim([bounds.miny.min()-5, bounds.maxy.max()+5]);
```

```python
# We can see that there are a few cities that intersect my path! But we want to know which ones.
# Do a spatial join to get the intersection
from geopandas.tools import sjoin
ecities = gpd.sjoin(cities, myeclipse, how='inner', op='intersects')
ecities.head()
```

```python
# Let's plot the results!
ax = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,5))
myeclipse.plot(ax=ax, color='yellow', edgecolor='black', alpha=0.5)
ecities.plot(marker='*', color='black', markersize=5, ax=ax)
bounds = myeclipse.geometry.bounds
plt.xlim([bounds.minx.min()-5, bounds.maxx.max()+5])
plt.ylim([bounds.miny.min()-5, bounds.maxy.max()+5]);
```

```python
ecities.head()
```

## Making a slippy map

```python
## Make it a slippy map
import folium
```

```python
# Let's find the centroid of the eclipse we chose, so that we can center our folum map on it
x = ecities.centroid.x.values[0]
y = ecities.centroid.y.values[0]
print('y: ' + str(y) + ', x: ' + str(x))
```

```python
map_carto = folium.Map(location=[y, x],tiles='Cartodb Positron',zoom_start=3)
map_carto
```

```python

folium.GeoJson(myeclipse,name='Eclipse Path').add_to(map_carto)

fg=folium.FeatureGroup(name="Pika")

for name,row in ecities.iterrows():
    popup_text = 'Location {}'.format(row['NAME'])
    lat = row['LATITUDE']
    lon = row['LONGITUDE']
    fg.add_child(folium.Marker(location=[lat,lon],popup=(folium.Popup(popup_text))))
    
map_carto.add_child(fg)
map_carto
```
